```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating count'})
# reset_index just reset index from 0 to so on
```

```python
ratings
```

# obervations
1. when rating is between 4.5 to 4.9 --> excellent
2. when rating is between 4.5 to 4.4 --> very good
3. when rating is between 3.5 to 3.9 --> good
4. when rating is between 3.0 to 3.4 --> average
5. when rating is between 2.5 to 2.9 --> average
6. when rating is between 2.0 to 2.4 --> poor

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating',y='Rating count',data=ratings,palette="pastel")
```

```python
sns.barplot(x='Aggregate rating',y='Rating count',data=ratings,hue='Rating color',palette=['black','red','orange','yellow','green','purple'])
```

```python
# obervations:-
# 1.not rated count is very high
# 2.maximum no of ratings are between 2.5 to 3.8
```

```python
# count plot
sns.countplot(x='Rating color',data=ratings,palette=['black','red','orange','yellow','green','purple'])
# this is frequncy not the rating count
```
